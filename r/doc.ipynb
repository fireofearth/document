```r
# plot columnar data carat and price in data frame as histogram using ggplot
plot1 <- ggplot(diamonds, aes(carat)) + geom_histogram(bins=60)
plot2 <- ggplot(diamonds, aes(price)) + geom_histogram(bins=60)
# set display option for plot
options(repr.plot.width=15, repr.plot.height=7)
# display grid of 2 plots
grid.arrange(plot1, plot2, ncol=2, widths = c(1,1), heights=1)
```
